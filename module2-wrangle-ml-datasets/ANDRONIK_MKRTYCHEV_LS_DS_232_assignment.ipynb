```python
# train, val and testz set
train = df[(df['Year'] >= 1550) & (df['Year'] <= 1870)]
val = df[(df['Year'] >= 1790) & (df['Year'] <= 1870)]
test = df[(df['Year'] >= 1795) & (df['Year'] <= 1870)]

# shape
train.shape, val.shape, test.shape
```

```python
# wrangle function
def wrangle(X):
  
  # Prevent SettingWithCopyWarning
  X = X.copy()
  
  return X
```

```python
# implement wrangle function
train = wrangle(train)
val = wrangle(val)
test = wrangle(test)
```

```python
# shape
train.shape, val.shape, test.shape
```

### Can you make a fast, first model that beats guessing?

```python
# feature selection
target = 'Real_consumption_earnings_Growth_GB'
features = train.columns.drop([target, 'Year'])
X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
```

```python
# import libraries
from numpy import nan
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
```

```python
# pipeline
pipeline = make_pipeline(
    SimpleImputer(missing_values=nan, strategy='median'),
    LinearRegression(n_jobs=5)
)
```

```python
# shape
X_train.shape, y_train.shape
```

```python
# fit model
pipeline.fit(X_train, y_train)
```

```python
# validation accuracy, looks like no leakage
print('Validation Accuracy', pipeline.score(X_val, y_val))
```
